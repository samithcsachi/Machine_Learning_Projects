## Importing the Libraries

```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings

warnings.filterwarnings('ignore')

from pycaret.classification import *
```

## Load the Dataset 

```python
df = pd.read_csv('/kaggle/input/breast-cancer-wisconsin-data/data.csv')
df.head()
```

```python
# Deleting the unwanted columns 
df = df.drop(columns = ['id','Unnamed: 32'], axis = 1,)
```

```python
#Basic Statistics 
df.describe()
```

```python
df.info()
```

```python
df['diagnosis'].value_counts()
```

## Exploratory Data Analysis 

```python
sns.countplot(x='diagnosis', data=df)
plt.show()
```

```python
df_temp =df.drop(columns=['diagnosis'], axis = 1 )
```

```python
#Create hist plot
fig, ax = plt.subplots(ncols = 6 , nrows= 5, figsize =(20,20))
index = 0
ax =ax.flatten()

for col in df_temp.columns: 
    sns.histplot(df[col], ax=ax[index], kde=True)
    index +=1
plt.tight_layout(pad =0.5, w_pad = 0.7, h_pad =5.0)
plt.show()
```

```python
#Create box plot
fig, ax = plt.subplots(ncols = 6 , nrows= 5, figsize =(20,20))
index = 0
ax =ax.flatten()

for col in df_temp.columns: 
    sns.boxplot(y =col, data=df, ax=ax[index])
    index +=1
plt.tight_layout(pad =0.5, w_pad = 0.7, h_pad =5.0)
plt.show()
```

## Creat and Train the Model 

```python
clf = setup(df, target='diagnosis')
```

```python
compare_models()
```

```python
# Select the best model 
model = create_model('catboost')
```

```python
# Hyperparameter Tuning 

best_model =tune_model(model)
```

```python
evaluate_model(best_model)
```

```python
# Plot the results 
plot_model(estimator= best_model, plot='confusion_matrix' )
```
